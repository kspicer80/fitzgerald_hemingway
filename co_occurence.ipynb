```python
import spacy
nlp = spacy.load('en_core_web_lg')

with open('data/hemingway/hemingway_sun_also.txt', 'r') as f:
    text = f.read()
```

```python
doc = nlp(text)

people = []

for entity in doc.ents:
    # Check if the entity label is "PERSON"
    if entity.label_ == "PERSON":
        people.append(entity.text)
```
